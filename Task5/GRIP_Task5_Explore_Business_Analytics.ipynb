```
#Checking Total Sales & Profit of the company
sales_profit = df.groupby('Segment').sum().iloc[:,[1,-1]].sum()
round(sales_profit,2)
```

```
pd.DataFrame.hist(df)
plt.show()
```

```
df.plot(kind ='density', subplots= True, layout=(3,2), sharex= False)
plt.show()
```

```
corr = df.corr()
sns.heatmap(corr,annot = True)
```

```
Top_Sales = df.groupby('State')['Sales'].sum().sort_values(ascending = False)
Top_Profit = df.groupby('State')['Profit'].sum().sort_values(ascending = False)
print(Top_Sales, Top_Profit)
```

```
print(Top_Sales.sum())
```

```
Top_Sales.head(10).plot(kind = 'bar',figsize = (15,6),fontsize = 14)
plt.title("Top 10 States by Sales",fontsize = 16)
plt.xlabel("States",fontsize = 13)
plt.ylabel("Sales",fontsize = 13)
```

MOST SALES IS IN STATE CALIFORNIA.
LEAST SALES IS IN STATE VIRGINIA.

```
Top_Profit.head(10).plot(kind = 'bar',figsize = (15,6),fontsize = 14)
plt.title("Top 10 States by Profit",fontsize = 16)
plt.xlabel("States",fontsize = 13)
plt.ylabel("Profit",fontsize = 13)
```

TOP 3 PROFIT STATES: CALIFORNIA, NEW YORK, WASHINGTON.

LEAST 3 PROFIT STATES: KENTUCKY, MINNESOTA, DELAWARE

```
plt.figure(figsize=(15,6))
sns.barplot('Region','Profit',hue = 'Segment',data = df)
plt.title("Regionwise Profit in Different Segment",fontsize = 13)
```

EAST has most profit is Segment in HOME OFFICE while others are low in the same.
WEST is moderate  in all segments
CENTRAL has least profit in CONSUMER SEGMENT

```
sns.pairplot(data=df,hue="Segment",palette="Set2")
plt.show()
```

MOST PROFITABLE SEGMMENTS ARE IN HOME OFFICES

```
central_region = list(df.groupby('Region'))[0][1]
plt.figure(figsize=(15,6))
sns.barplot('Category','Profit',hue = 'Ship Mode',data = central_region)
plt.title('Categorywise Profit in Central Region through Ship Mode',fontsize = 13)
```

```
central_region_furniture = list(list(df.groupby("Region"))[0][1].groupby("Category"))[0][1]
plt.figure(figsize=(15,6))
sns.barplot('State',"Profit", hue = 'Segment',data = central_region_furniture)
```

```
plt.figure(figsize=(15,6))
sns.barplot('State','Discount',hue ='Sub-Category',data = central_region_furniture )
```

The discounts are are mostly offered in Texas and illinois wherein discounts are provided at the below percentages:
a)Most provided discount is 60% on Furnishings
b)30% on Bookcases and Chairs
c)Illinois provides more discount on Tables than Texas
c)50% disount on Tables in Illinois and 30% in Texas
